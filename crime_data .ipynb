# Hierarchical clustering

```python
import pandas as pd
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv("E:\\data science\\assiments\\7\\crime_data.csv")
```

```python
data
```

```python
array=data.iloc[:,1:]
```

```python
array
```

```python
stscaler=StandardScaler().fit(array)
```

```python
x=stscaler.transform(array)
```

```python
x
```

```python
dendrogram = sch.dendrogram(sch.linkage(x, method='complete'))
```

```python
hc = AgglomerativeClustering(n_clusters=2, affinity = 'euclidean', linkage = 'complete')
hc
```

```python
y_hc = hc.fit_predict(x)
y_hc
```

```python
data['clusterid']=pd.Series(y_hc)
```

```python
data.head()
```

```python
result=data.iloc[:,1:].groupby(data.clusterid).mean()
```

```python
result
```

# K-Means Clustering

```python
import pandas as pd
import matplotlib.pylab as plt
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist 
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
data1=pd.read_csv("E:\\data science\\assiments\\7\\crime_data.csv")
```

```python
data1.head()
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
x1 = norm_func(data1.iloc[:,1:])
x1.head()
```

```python
kmeans_clust=KMeans(n_clusters=5)
kmeans_clust.fit(x1)
```

```python
kmeans_clust.labels_
```

```python
data1['clusterid']=pd.Series(kmeans_clust.labels_)
```

```python
data1
```

```python
result=data1.groupby(data1.clusterid).mean()
```

```python
result
```

```python
kmeans_clust.cluster_centers_
```

```python
kmeans_clust.cluster_centers_[0].shape
```

```python
k = list(range(2,11))
k
TWSS = [] 
for i in k:
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(x1)
    WSS = [] # variable for storing within sum of squares for each cluster 
    for j in range(i):
        WSS.append(sum(cdist(x1.iloc[kmeans.labels_==j,:],kmeans.cluster_centers_[j].reshape(1,x1.shape[1]),"euclidean")))
        
    print("TWSS when k is",i,sum(WSS))
    TWSS.append(sum(WSS))
    
```

```python
plt.plot(k,TWSS, 'ro-');
plt.xlabel("No_of_Clusters");
plt.ylabel("total_within_SS")
```
